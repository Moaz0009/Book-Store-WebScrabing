```python
from bs4 import BeautifulSoup 
import requests 
import csv
```

```python
site = requests.get("https://books.toscrape.com/catalogue/page-1.html")
site
```

```python
pages=[1]
data={'Title':[],'price':[],'avilable':[],'img_link':[]}
for page in pages:
    site = requests.get(f"https://books.toscrape.com/catalogue/page-{page}.html").text
    website = BeautifulSoup(site, "lxml")
    books=website.find_all('li',class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
    for book in books:
        data['Title'].append((book.find('h3')).find('a')['title'])
        data['price'].append(book.find('p',class_='price_color').text)
        avail=book.find('p',class_='instock availability').text
        data['avilable'].append(avail.strip())
        img_tag=book.find('a')
        img_src=img_tag.find('img')
        data['img_link'].append(img_src['src'])
```

```python
import pandas as pd
df=pd.DataFrame(data)
df
```
